```python
import warnings
warnings.filterwarnings('ignore')
import openpyxl
import numpy as np
from Expolynomial import *
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
p = np.array([])
n = -3
while n <= 12:
    p = np.append(p,[n])
    n = n + 0.02
    
p1 = np.array([])
n = -5
while n <= 12:
    p1 = np.append(p1,[n])
    n = n + 0.04
    
```

```python
wb = openpyxl.load_workbook('framingham.xlsx',data_only=True)
```

```python

bin_tot = np.array([])
s = -0.7825
while(s<1.3):
    bin_tot = np.append(bin_tot,[s])
    s = s + 0.10986
    
bin_glu = np.array([])
s = -0.6678
while(s<1.86):
    bin_glu = np.append(bin_glu,[s])
    s = s + 0.1334
"""
bin_tot = np.array([])
s = -6
while(s<8):
    bin_tot = np.append(bin_tot,[s])
    s = s + 0.5
    
bin_glu = np.array([])
s = -6
while(s<8):
    bin_glu = np.append(bin_glu,[s])
    s = s + 0.5
"""
    
```

```python
Binlist = [[],np.array(range(40,180,10)),np.array(range(70,325,15)),np.array(range(15,59,2)),np.array(range(40,152,4)),bin_tot,bin_glu,np.array(range(32,74,3)),[]]
```

```python
for sheet,bins in zip(wb.worksheets,Binlist):
    print("No Stroke: {0}".format(sheet.title))
    if(sheet.title != "Framingham Data" and sheet.title != "cigsPerDay"):
        da = np.array([])
        for rowOfCellObjects in sheet['C7':'C4250']:        
            da = np.append(da,[rowOfCellObjects[0].value])
        da = da[da != np.array(None)]
        da = da[da != np.array('NA')]
        da = da.astype(float)
        #sns.distplot(da,kde_kws={"label": "KDE"})
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            ex = expolynomial_pdf(da,bins)
            result = ex.evaluate(p1,space = "lny")
        else:
            ex = expolynomial_pdf(da,bins)
            result = ex.evaluate(p)
        
        pdfx = result[0]
        fit = result[1]
        m = np.median(da)
        s = np.sqrt(np.mean(abs(da - m)**2))
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            dataset_lny = np.array([])
            for e in da:
                dataset_lny = np.append(dataset_lny,[math.log(e/float(m))])
            lny_median = np.median(dataset_lny)           
            lny_std = np.sqrt(np.mean(abs(dataset_lny - lny_median)**2))
            p_za = p1 * lny_std
            p_y = exp(p_za)
            pointx = p_y*m
            
            log1 = pointx >= min(da) *0.7
            pointx = pointx[log1]
            pdfx = pdfx[log1]
            log2 = pointx <= max(da)*1.1
            pointx = pointx[log2]
            pdfx = pdfx[log2]
        else:
            pointx = (p * s/float(m) + 1) * m
            
            log1 = pointx >= min(bins) * 0.6
            pointx = pointx[log1]
            pdfx = pdfx[log1]
            log2 = pointx <= max(bins) 
            pointx = pointx[log2]
            pdfx = pdfx[log2]
        print("x: {0}".format(fit.x))
        print("message: {0}".format(fit.message))
        print("nfev: {0}".format(fit.nfev))
        
        if(type(bins) == int):
            print("MSE: {0}".format(fit.cost/bins))
        else:
            print("MSE: {0}".format(fit.cost/len(bins)))
        
        sns.distplot(da,kde_kws={"color": "b", "lw": 1, "label": "KDE"})
        plt.plot(pointx,pdfx)
        
        #plt.xlim(left = min(bins),right = max(bins))
        plt.show()
        print('--- END OF ROW ---')
        
```

```python
for sheet,bins in zip(wb.worksheets,Binlist):
    print("Stroke: {0}".format(sheet.title))
    if(sheet.title != "Framingham Data" and sheet.title != "cigsPerDay"):
        da = np.array([])
        for rowOfCellObjects in sheet['E7':'E35']:        
            da = np.append(da,[rowOfCellObjects[0].value])
        da = da[da != np.array(None)]
        da = da[da != np.array('NA')]
        da = da.astype(float)
        #sns.distplot(da,kde_kws={"label": "KDE"})
        binx = np.array([])
        binh = np.array([])
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            for rowOfCellObjects in sheet['P6':'P40']:        
                binx = np.append(binx,[rowOfCellObjects[0].value])
            binx = binx[binx != np.array(None)]
            binx = binx.astype(float)
        
            
            for rowOfCellObjects in sheet['V6':'V40']:        
                binh = np.append(binh,[rowOfCellObjects[0].value])
            binh = binh[binh != np.array(None)]
            binh = binh.astype(float)
            
        else:
            
            for rowOfCellObjects in sheet['N6':'N40']:        
                binx = np.append(binx,[rowOfCellObjects[0].value])
            binx = binx[binx != np.array(None)]
            binx = binx.astype(float)
        
            
            for rowOfCellObjects in sheet['T6':'T40']:        
                binh = np.append(binh,[rowOfCellObjects[0].value])
            binh = binh[binh != np.array(None)]
            binh = binh.astype(float)
        
        
        insert_cm = np.array([])
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            for rowOfCellObjects in sheet['U7':'U28']:
                insert_cm=np.append(insert_cm,[rowOfCellObjects[0].value])
            
        else:
            for rowOfCellObjects in sheet['S7':'S36']:
                insert_cm=np.append(insert_cm,[rowOfCellObjects[0].value])

        insert_cm = insert_cm[insert_cm != np.array(None)]
        insert_cm = insert_cm[insert_cm != np.array('NA')]
        insert_cm = insert_cm.astype(float)
        insert_cm = insert_cm.tolist()
        #print(insert_cm)
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            ex = expolynomial_pdf(da,bins,order = 4)
            result = ex.evaluate(p1,space = "lny",insert_cmf = insert_cm)
        else:
            ex = expolynomial_pdf(da,bins,order = 3)
            result = ex.evaluate(p,insert_cmf = insert_cm)
        
        pdfx = result[0]
        fit = result[1]
        m = np.median(da)
        s = np.sqrt(np.mean(abs(da - m)**2))
        if(sheet.title == "totChol" or sheet.title == "glucose"):
            dataset_lny = np.array([])
            for e in da:
                dataset_lny = np.append(dataset_lny,[math.log(e/float(m))])
            lny_median = np.median(dataset_lny)
           
            lny_std = np.sqrt(np.mean(abs(dataset_lny - lny_median)**2))
            
            p_za = p1 * lny_std
            p_y = exp(p_za)
            pointx = p_za
            resultz = ex.pdfz(fit.x,p1)
            pdfx = resultz/lny_std
            
            log1 = pointx >= min(bins)
            pointx = pointx[log1]
            pdfx = pdfx[log1]
            log2 = pointx <= max(bins)
            pointx = pointx[log2]
            pdfx = pdfx[log2]
        else:
            pointx = (p * s/float(m) + 1) * m
            
            log1 = pointx >= min(bins) * 0.6
            pointx = pointx[log1]
            pdfx = pdfx[log1]
            log2 = pointx <= max(bins) 
            pointx = pointx[log2]
            pdfx = pdfx[log2]
        print("x: {0}".format(fit.x))
        print("message: {0}".format(fit.message))
        print("nfev: {0}".format(fit.nfev))
       
        if(type(bins) == int):
            print("MSE: {0}".format(fit.cost/bins))
        else:
            print("MSE: {0}".format(fit.cost/len(bins)))
        
        #logic = (pointx >= min(bins) and pointx <= max(bins))
        binwidth = bins[1]-bins[0]
        plt.bar(binx,binh,width = binwidth,align = 'edge')
        #print(binx)
        #print(binh)
        plt.plot(pointx,pdfx,color='orange')
        
        
        #sns.distplot(da,kde_kws={"color": "b", "lw": 1, "label": "KDE"})
        
        #plt.xlim(left = min(da)*0.9,right = max(da)*1.1)
        #plt.ylim(bottom = 0,top = 0.08)
        
        plt.show()
        print('--- END OF ROW ---')
```
